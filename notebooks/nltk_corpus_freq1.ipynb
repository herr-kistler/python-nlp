**1. How many words in the Chat Corpus (`text5`) start with the letter *z*?**

This question is ambiguous: Does "words" could mean either types or tokens? And does case matter?

`````python
from nltk.book import *
`````

`````python
# tokens, case sensitive

z = [w for w in text5 if w.startswith('z')]
print(len(z), z)
`````

`````python
# tokens, case insensitive

z = [w for w in text5 if w.lower().startswith('z')]
print(len(z), z)
`````

**2. What is the difference between the following two lines? Which one will give a larger value? Will this be the case for other texts?**
````
>>> sorted(set(w.lower() for w in text1))
>>> sorted(w.lower() for w in set(text1))
````

First compare the lengths:

`````python
len(sorted(set(w.lower() for w in text1))), len(sorted(w.lower() for w in set(text1)))
`````

Why the difference?  The first one gets rid of case differences and then finds the unique items (i.e., the types).  The second finds the types and then gets ride of case differences.  That means that two types the differ in case will be counted once by the first expression and twice by the second.  This is going to be true for all texts, as we can see with a small example:

`````python
len(sorted(set(w.lower() for w in ['The', 'the', 'THE']))), len(sorted(w.lower() for w in set(['The', 'the', 'THE'])))
`````

** 4. Read in the texts of the State of the Union addresses, using the `state_union` corpus reader. Count occurrences of *men*, *women*, and *people* in each document. What has happened to the usage of these words over time? **





*People* has always been more frequent that the others, but *men* was more common than *women* up to the 1970's.  Since then, *men* and *women* have been (roughly) equally frequent.

`````python
from nltk import ConditionalFreqDist
from nltk.corpus import state_union

%matplotlib inline
`````

`````python
sotu_cfd = ConditionalFreqDist([(w.lower(),name[:4]) for name in state_union.fileids()
                                                     for w in state_union.words(name)
                                                     if w.lower() in ['men', 'women', 'people']])
sotu_cfd.plot()
`````

NLTK's default plot settings are usually okay, but they don't work very well at all for this graph.  We can get get much better-looking results by calling [matplotlib/pylab](http://matplotlib.org/) directly, like this:

`````python
import matplotlib.pylab as plt

years = sorted(set([y for word in sotu_cfd.conditions()
               for y in sotu_cfd[word].keys()]))

for word in sotu_cfd.conditions():
    plt.plot([sotu_cfd[word][year] for year in years], label=word, linewidth=2)

plt.grid(True, color='silver')
plt.legend(loc='upper left')
plt.xticks(range(0,len(years),4), years[::4], rotation=90)   # only label every 4th year
plt.xlabel('Year')
plt.ylabel('Counts')

plt.show()
`````

** 5. Write a program to find all words that occur at least 100 times in the Brown Corpus.**


`````python
from nltk.corpus import brown
from nltk import FreqDist
`````

`````python
brown_fd = FreqDist([w.lower() for w in brown.words() if w.isalpha()])
[w for w in brown_fd if brown_fd[w] >= 100]
`````

**6. Write a program to guess the number of syllables contained in a text, making use of the CMU Pronouncing Dictionary. [This one is hard!  If you haven't done much programming before, you'll probably find it very challenging, but try your best.]**

`````python
from nltk.corpus import cmudict
`````

The function `stress()` on page 64 is a good place to start:

`````python
def stress(pron):
    return [char for phone in pron for char in phone if char.isdigit()]
`````

The length of the stress pattern tells us the number of syllables in a pronunciation.  All that's left to do is to find the pronunciation of a word.  Looking up a word in the cmudict dictionary returns a list of possible pronunciations:

`````python
pron = cmudict.dict()
pron['fire']
`````

To keep things simple, we'll just take the first pronunciation on the list, pass it to `stress()` to to get the stress pattern, and then take the length of that:

`````python
print(pron['fire'])
print(pron['fire'][0])
print(stress(pron['fire'][0]))
print(len(stress(pron['fire'][0])))
`````

Finally, we'll put it all together into a function:

`````python
def sylls(text):
    text = [w.lower() for w in text if w.isalpha()]
    sylls = 0
    for word in text:
        if word in pron:
            sylls += len(stress(pron[word][0]))
    return sylls
`````

`````python
sylls('This is a sentence too'.split())
`````

The function `sylls()` skips any word that isn't in the CMU dictionary:

`````python
sylls(['Nicki', 'Minaj'])
`````

An improvement might be to guess at the syllable count for unlisted words by counting the vowel groups:

`````python
import re
`````

`````python
def sylls2(text):
    text = [w.lower() for w in text if w.isalpha()]
    syll = 0
    for word in text:
        if word in pron:
            syll += len(stress(pron[word][0]))
        else:
            syll += len(re.findall(r'[aeiou]+',word))
    return syll        
`````

`````python
sylls2(['Nicki', 'Minaj'])
`````

Note that this will go wrong for words that end in a silent *e*:

`````python
sylls2(['eLance'])
`````

To fix this, we can check for a silent e when we count vowel groups and subtract one from the syllable count if we find one:

`````python
def sylls3(text):
    text = [w.lower() for w in text if w.isalpha()]
    syll = 0
    for word in text:
        if word in pron:
            syll += len(stress(pron[word][0]))
        else:
            syll += len(re.findall(r'[aeiou]+',word))
            if re.search(r'[aeiou][^aeiou]+e$',word):
                syll -= 1
    return syll    
`````

`````python
sylls3(['eLance'])
`````
